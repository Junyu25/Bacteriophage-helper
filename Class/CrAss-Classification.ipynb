```python
crassAnno = []
for seq in SeqIO.parse(crass, "fasta"):
    for i in range(len(df)):
        if df["Unnamed: 0"][i] in seq.id.lower():
            #seq.id = seq.id + "-" + str(df["clusters_kopt"][i]) + "-" + df["Topology"][i]
            seq.id = seq.id.upper()
            seq.description = df["Topology"][i] + "-cluster" + str(df["clusters_kopt"][i])
            crassAnno.append(seq)

SeqIO.write(crassAnno, "/home/junyuchen/Lab/Phage-SOP/CraAss-Classification/crass_genome_Anno-249-1.fasta" , "fasta")    
```
