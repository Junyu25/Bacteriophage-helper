```python
Desc = df.ConsensusFunctionalDescription
for i in range(len(df)):
    hmm = os.path.join(Dir, df["#GroupName"][i]+".hmm")
    #print(hmm)
    if os.path.exists(hmm):
        #print(hmm)
        with open(hmm, "r") as in_file:
            buf = in_file.readlines()

        with open(hmm, "w") as out_file:
            for line in buf:
                if line == "NAME  " + df["#GroupName"][i] + "\n":
                    #print(line)
                    line = line + "DESC  " + Desc[i].split("|")[-1] + "\n"
                out_file.write(line)
```
